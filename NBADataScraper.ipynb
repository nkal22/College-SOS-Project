```python
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

import numpy as np
from datascience import *

# These lines do some fancy plotting magic.\n",
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')

```

```python
dataTable = pd.read_csv (r'/Users/nicholaskalinowski/Documents/NBADraftData.csv', encoding= 'unicode_escape')

dataTable['sos'] = 0

#dataTable.drop(dataTable[dataTable['College'] == 'None'].index, inplace = True)

dataTable = dataTable.loc[dataTable["College"] != 'None']
dataTable = dataTable.reset_index()

index1 = 0


for best_player in dataTable.Player:
    best_player_2 = best_player.replace(' ', '-')
    best_player_url = best_player_2 + "-1"
    best_player_url = best_player_url.lower()
    best_player_url = best_player_url.replace('.', '')
    best_player_url = best_player_url.replace("'", '')
    best_player_url = best_player_url.replace('-jr', 'jr')
    best_player_url = best_player_url.replace('-iii', 'iii')
    best_player_url = best_player_url.replace('devyn', 'roy-devyn')
    if best_player_url == 'patrick-williams-1':
        best_player_url = 'patrick-williams-2'
    if best_player_url == 'obi-toppin-1':
        best_player_url = 'obadiah-toppin-1'
    if best_player_url == 'anthony-edwards-1':
        best_player_url = 'anthony-edwards-2'
    if best_player_url == 'nikola-vuevi-1':
        best_player_url = 'nikola-vucevic-1'
    if best_player_url == 'josh-green-1':
        best_player_url = 'josh-green-2'
    if best_player_url == 'xavier-tillman-sr-1':
        best_player_url = 'xavier-tillman-1'
    if best_player_url == 'nick-richards-1':
        best_player_url = 'nick-richards-2'
    if best_player_url == 'reggie-perry-1':
        best_player_url = 'reggie-perry-2'
    if best_player_url == 'ja-morant-1':
        best_player_url = 'temetrius-morant-1'
    if best_player_url == 'kz-okpala-1':
        best_player_url = 'kezie-okpala-1'
    if best_player_url == 'dewan-hernandez-1':
        best_player_url = 'dewan-huell-1'
    if best_player_url == 'mo-bamba-1':
        best_player_url = 'mohamed-bamba-1'
    if best_player_url == 'lonnie-walker-1':
        best_player_url = 'lonnie-walker-2'
    if best_player_url == 'shake-milton-1':
        best_player_url = 'malik-milton-1'
    if best_player_url == 'ray-spalding-1':
        best_player_url = 'raymond-spalding-1'
    if best_player_url == 'bam-adebayo-1':
        best_player_url = 'edrice-adebayo-1'
    if best_player_url == 'kay-felder-1':
        best_player_url = 'kahlil-felder-1'
    if best_player_url == 'larry-nancejr-1':
        best_player_url = 'larry-nance-2'
    if best_player_url == 'tim-hardawayjr-1':
        best_player_url = 'tim-hardaway-jr-1'
    if best_player_url == 'glen-ricejr-1':
        best_player_url = 'glen-rice-jr-1'
    if best_player_url == 'ray-mccallum-1':
        best_player_url = 'ray-mccallum-2'
    if best_player_url == 'thomas-robinson-1':
        best_player_url = 'thomas-robinson-2'
    if best_player_url == 'bernard-james-1':
        best_player_url = 'bernard-james--1'
    if best_player_url == 'kyle-anderson-1':
        best_player_url = 'kyle-anderson-3'
    url = "https://www.sports-reference.com/cbb/players/" + best_player_url + ".html"
    html = urlopen(url)
    soup = BeautifulSoup(html)
    #soup.findAll('tr', limit=2)
    headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
    headers = headers[28:]
    #headers
    rows = soup.findAll('tr')[-1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
                    for i in range(len(rows))]
    #print(player_stats[0])
    stats = pd.DataFrame(player_stats, columns=('A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z', 'aa', 'sos'))
    stats = stats.drop(['A','B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', "w", 'x', 'y', 'z', 'aa'], axis=1)
    dataTable.loc[index1, 'sos'] = stats.sos[0]
    index1 = index1 + 1
    
dataTable

```
